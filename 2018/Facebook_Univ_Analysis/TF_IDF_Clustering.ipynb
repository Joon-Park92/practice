### Prediction by similarity Bag of words vector with Euclidian distance

```python
def dist_norm(v1,v2):
    v1_normalized = v1 /sp.linalg.norm(v1)
    v2_normalized = v2 /sp.linalg.norm(v2)
    delta = v1_normalized - v2_normalized
    
    if sp.linalg.norm(delta) == 0 :
        re = 0
    else:
        re = sp.linalg.norm(delta)    
    return re
```
